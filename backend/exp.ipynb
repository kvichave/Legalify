```python
import pymupdf # previously known as fitz

doc = pymupdf.open(r"C:\Users\DNFH3173\Desktop\Legalify\temp\master_agreement.pdf")
for i, page in enumerate(doc):
    # Extract page content as HTML
    html_content = page.get_text("html")

    # Save each page to a separate file
    with open(f"page-{i}.html", "w", encoding="utf-8") as f:
        f.write(html_content)
doc.close()
```
