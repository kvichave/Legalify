```python
import pymupdf # previously known as fitz

doc = pymupdf.open(r"C:\Users\DNFH3173\Desktop\Legalify\temp\master_agreement.pdf")
for i, page in enumerate(doc):
    # Extract page content as HTML
    html_content = page.get_text("html")

    # Save each page to a separate file
    with open(f"page-{i}.html", "w", encoding="utf-8") as f:
        f.write(html_content)
doc.close()
```

```python
import pymupdf4llm
import markdown

# Extract to Markdown
md_text = pymupdf4llm.to_markdown(r"C:\Users\DNFH3173\Desktop\Legalify\temp\invoice.pdf")
# Convert to HTML
html_content = markdown.markdown(md_text)
```

```python
from langchain_openai import ChatOpenAI
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
llm = ChatOpenAI(
    model="DeepSeek-V3.2",
    base_url="https://agentic-p.openai.azure.com/openai/v1",
    api_key=os.getenv("AZURE_MODEL"),
    temperature=0,
)
```

```python
llm.invoke([{"role": "user", "content": "What is the capital of France?"}])
```
